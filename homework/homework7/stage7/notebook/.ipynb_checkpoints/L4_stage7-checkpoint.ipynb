## Stage 7: Outliers + Risk Assumptions



```python
# 1) Setup
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

# Raw Data
np.random.seed(7)
x = np.linspace(0, 10, 60)
y = 1.8 * x + 0.5 + np.random.normal(0, 1.8, size=x.size)
# Inject a few extreme points
y[8] += 20
y[25] -= 18
y[47] += 15


#creating a table for the data (x,y)
df = pd.DataFrame({'x': x, 'y': y})
df.head()
```

```python
#Data Visualization - scatterplot

plt.figure()
plt.scatter(df['x'], df['y']) #select data frame
plt.title('Scatter with Potential Outliers')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## 3) Detecting Outliers — IQR
Range: (Q1 - 1.5 * k , Q3 + 1.5 * k)

k = Q3 - Q1

```python
def detect_outliers_iqr(series: pd.Series, k: float = 1.5) -> pd.Series:
    """Return boolean mask where True indicates an outlier by IQR rule. 
    
    Parameters:
    ----------
    series : pd.Series - Numeric series to evaluate/the numbers
    k : float - Multiplier for IQR to set fences (default 1.5).
    """
    
    q1 = series.quantile(0.25) 
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - k * iqr
    upper = q3 + k * iqr
    return (series < lower) | (series > upper) 
    #T = within IQR
    #F = outside IQR

df['outlier_iqr'] = detect_outliers_iqr(df['y']) #T/F
df['outlier_iqr'].value_counts() #returns number of T/F exist 


df['outlier_iqr'] = detect_outliers_iqr(df['y'],.6)
df.head(6)
df.shape[0],df['outlier_iqr'].value_counts() 

#df.shape(row,column) - returns number of row and column
#df.shape[0] - index position - in this case, only returns number of rows
```

```python
# Box Plots

plt.figure()
plt.boxplot(df['y'], vert=True)
plt.title('Boxplot of y (with potential outliers)')
plt.ylabel('y')
plt.show()
```

```python
#Removing Outliers - bothside has extremes
#Recognize both min and max, and removing them 

df2 = df.copy()
max_mask = (df2['y'] == max(df2['y']))
min_mask = (df2['y'] == min(df2['y']))
mask = max_mask | min_mask

plt.figure()
plt.boxplot(df['y'][-mask], vert=True)
plt.title('Boxplot of y (with potential outliers)')
plt.ylabel('y')
plt.show()
```

## 4) Detecting Outliers — Z-score

Flag Z-Score up to some value

```python
# Ternary operator
sazszaxxazsazs
```

pd.Series.mean(axis=0, skipna=True, numeric_only=False, **kwargs)

**Axis**
        # 0 = mean of each column (default)
        # 1 = mean of each row
        
**skipna** = skips NaNs, not considered into the formula

**numeric_only** = False - if the Series has mixed types, default = False, use all

**numeric_only** =  True -->  will only use numeric values

```python
def detect_outliers_zscore(series: pd.Series, threshold: float = 3.0) -> pd.Series:
    """Return boolean mask where True indicates |z| > threshold."""
    
    mu = series.mean()
    sigma = series.std(ddof=0)
    z = (series - mu) / (sigma if sigma != 0 else 1.0)
    return z.abs() > threshold

df['outlier_z'] = detect_outliers_zscore(df['y'], threshold=3.0)
# threshold=3.0) - within 3 stdev
# detect_outliers_zscore find 1 z-score


pd.crosstab(df['outlier_iqr'], df['outlier_z'])
#pd.crosstab(index, columns) - returns a frequency table
```

## 5) Handling Options
how to deal with outliers

We will compare three treatments: **All data**, **Filtered (remove IQR outliers)**, and **Winsorized** (optional).

```python
def winsorize_series(series: pd.Series, lower: float = 0.05, upper: float = 0.95) -> pd.Series:
    """Return a winsorized copy of the series, clipping to given quantiles."""
    lo = series.quantile(lower)
    hi = series.quantile(upper)
    return series.clip(lower=lo, upper=hi)

df_filtered = df.loc[~df['outlier_iqr'], ['x', 'y']].reset_index(drop=True)
df_w = df.copy()
df_w['y_w'] = winsorize_series(df_w['y'], lower=0.05, upper=0.95)

#Winsorization - pulls outlier into the boundary
```

```python
plt.figure()
plt.hist(df_w['y_w'], bins=20, alpha=0.7)
plt.title('Histogram of y (winsorized)')
plt.xlabel('y_w')
plt.ylabel('count')
plt.show()
```
